```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from pandas_ods_reader import read_ods
from sklearn import linear_model
```

```python
sheetidx = 1
path = "/home/iamchiranjeeb/Desktop/amita/canada_per_capita_income.ods"
df = read_ods(path,sheetidx)
df
```

```python
plt.scatter(df.year,df.capita_income,color='red',marker="+",label="Scatter Plot")
plt.title("Canada Per Capita Income",fontsize=10)
plt.xlabel("Year -> ",fontsize=15)
plt.ylabel("Per Capita Income -> ",fontsize=15)
plt.legend()
plt.show()
```

```python
X = df[["year"]]
X
```

```python
y=df.capita_income
y
```

```python
reg = linear_model.LinearRegression()
```

```python
reg.fit(X,y)
```

```python
prediction = reg.predict([[2020]])
prediction
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
plt.scatter(df.year,df.capita_income,color='red',marker="+",label="Scatter Plot")
plt.plot(df.year,reg.predict(df[["year"]]),color='blue',label="Regression Line")
plt.title("Canada Per Capita Income",fontsize=10)
plt.xlabel("Year -> ",fontsize=15)
plt.ylabel("Per Capita Income -> ",fontsize=15)
plt.legend()
plt.show()
```
